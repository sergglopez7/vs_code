```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
x = np.random.rand(50)   # 50 números entre 0 y 1
y = np.random.rand(50)

# Crear gráfico de dispersión
plt.scatter(x, y)

# Títulos
plt.title("Gráfico de dispersión aleatorio")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Mostrar gráfico
plt.show()
```
